## Import Matplotlib, Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Load Built-in datasets from Seaborn

```python
penguins_df = sns.load_dataset("penguins")
flights = sns.load_dataset("flights")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
planets = sns.load_dataset("planets")
```

## Line Graph

## Docs: https://seaborn.pydata.org/generated/seaborn.lineplot.html

```python
# Handmade sample data
study_time = [1, 2, 3, 4, 5]  # Independent variable: study time (hours)
test_scores = [45, 70, 45, 80, 85]  # Dependent variable: test scores (%)

# Plotting the line graph
plt.plot(study_time, test_scores, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Study Time (hours)')
plt.ylabel('Test Scores (%)')
plt.title('Study Time vs. Test Scores')

# Display the plot & designing grid lines
plt.grid(True)
plt.show()
```

## Histogram

## Docs: https://seaborn.pydata.org/generated/seaborn.histplot.html

```python
sns.histplot(penguins_df, x="flipper_length_mm")
plt.xlabel("Flipper Length") # Have to use Matplotlib to customize it
```

We can also change the details of a histogram by passing in different arguments.

For instance, if we want to change how many bins there are, we can change `binwidth`, which controls the width of an individual bin, or change `bins` which directly determines how many bins the histogram has.

```python
sns.displot(penguins_df, x="flipper_length_mm", binwidth=3)
sns.displot(penguins_df, x="flipper_length_mm", bins=20)
```

```python
sns.histplot(data=penguins_df, x="flipper_length_mm", kde=True)
```

## Scatter Plot

## Docs: https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
# Generating random data using Numpy
x = np.random.rand(50)
y = np.random.rand(50)

# Creating scatter plot
plt.scatter(x, y,)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')

# Displaying the plot
plt.show()
```

## Box Plot

## Docs: https://seaborn.pydata.org/generated/seaborn.boxplot.html

### How to Read a Box Plot

image.png


```python
iris.head() # Iris dataset
```

```python
# x - should be a categorical variable
# y - should be numerical values
sns.boxplot(
    x='species',
    y='sepal_length',
    data=iris,
    showmeans=True    # use to show where mean is (green triangle)
    )
plt.xlabel('Species')  # customizing the x-axis label
plt.ylabel('Sepal Length (cm)')  # customizing the y-axis label
plt.title('Sepal Length Distribution by Species')  # adding a title
plt.show()
```
